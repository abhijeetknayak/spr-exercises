```python
data = np.load('../data/dataset.npz', mmap_mode='r')['arr_0']
data.shape
plt.scatter(data[:, 0], data[:, 1], c=data[:, 2])
plt.show()
```

```python
train_data, test_data = train_test_split(data, test_size=0.3)
print(train_data.shape, test_data.shape)
```

```python
# Linear SVM for Multiclass Classification
linear_svm = svmutil.svm_train(train_data[:, 2], train_data[:, 0:2], '-s 0 -t 0')
```

```python
p_labels, p_acc, p_vals = svmutil.svm_predict(test_data[:, 2], test_data[:, 0:2], linear_svm)
```

```python
plt.scatter(train_data[:, 0], train_data[:, 1], c=train_data[:, 2])
plt.scatter(test_data[:, 0], test_data[:, 1], c=p_labels, marker='*', s=100)
```

```python
# RBF Kernel SVM
rbf_svm = svmutil.svm_train(train_data[:, 2], train_data[:, 0:2], '-s 0 -t 2')
```

```python
p_labels, p_acc, p_vals = svmutil.svm_predict(test_data[:, 2], test_data[:, 0:2], rbf_svm)
```

```python
plt.scatter(train_data[:, 0], train_data[:, 1], c=train_data[:, 2])
plt.scatter(test_data[:, 0], test_data[:, 1], c=p_labels, marker='*', s=100)
```
